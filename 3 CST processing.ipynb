# CST (Central Subfield Thickness) Data Processing

**Objective**: Process raw CST.csv to match the format of BCVA_processed.csv

**Input**: `CST.csv` (raw data with multi-level headers)

**Output**: `CST_processed.csv` with clean column names

```python
import pandas as pd
import numpy as np

# Display settings
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', None)
```

## 1. Load Raw Data

```python
# Load raw CST data
cst_raw = pd.read_csv(r'c:\Users\ronny\Documents\GitHub\Longitudinal Prediction\raw data\CST.csv')

print(f"Raw data shape: {cst_raw.shape}")
cst_raw
```

```python
# Check the header structure
print("Row 0 (column names from file):")
print(cst_raw.columns.tolist()[:15])
print("\nRow 0 values (actual header info):")
print(cst_raw.iloc[0, :15].tolist())
```

## 2. Parse Header Structure

The CSV has a two-row header:
- Row 0: Time points (Screen, Week 4, Week 8, ...)
- Row 1: Variable names (Patient ID, Eye, Arm, CST, CST, Change, ...)

```python
# Reload with proper handling - skip the first row which contains time point info
cst_raw = pd.read_csv(r'c:\Users\ronny\Documents\GitHub\Longitudinal Prediction\raw data\CST.csv', skiprows=1)

print(f"Shape after skipping header row: {cst_raw.shape}")
print(f"\nColumns: {cst_raw.columns.tolist()[:15]}")
cst_raw.head(3)
```

## 3. Define Column Mapping

Map raw columns to clean names matching BCVA_processed.csv format

```python
# Define time points
time_points = ['Screen', 'Week4', 'Week8', 'Week12', 'Week16', 'Week20', 'Week24', 
               'Week28', 'Week32', 'Week36', 'Week40', 'Week44', 'Week48', 'Week52',
               'Week60', 'Week68', 'Week76', 'Week84', 'Week92', 'Week100', 'Week104']

# Build column name mapping
# Pattern: Patient ID, Eye, Arm, CST (Screen), CST (W4), Change (W4), CST (W8), Change (W8), ...

new_columns = ['Patient_ID', 'Eye', 'Arm', 'Screen']  # First 4 columns

# For Week4 onwards: CST value + Change
for tp in time_points[1:]:  # Skip 'Screen' as it has no change column
    new_columns.append(tp)           # CST value
    new_columns.append(f'{tp}_Change')  # Change from baseline

print(f"Number of new column names: {len(new_columns)}")
print(f"Number of raw columns: {len(cst_raw.columns)}")
print(f"\nNew column names: {new_columns[:15]}...")
```

```python
# Verify column count matches
if len(new_columns) == len(cst_raw.columns):
    print(" Column count matches!")
else:
    print(f" Mismatch: {len(new_columns)} new names vs {len(cst_raw.columns)} raw columns")
    print(f"  Difference: {len(cst_raw.columns) - len(new_columns)}")
```

```python
# Apply new column names
cst_processed = cst_raw.copy()
cst_processed.columns = new_columns

print("Processed columns:")
print(cst_processed.columns.tolist())
```

## 4. Data Validation

```python
# Check basic info
print(f"Number of patients/eyes: {len(cst_processed)}")
print(f"Unique Patient IDs: {cst_processed['Patient_ID'].nunique()}")
print(f"\nEye distribution:")
print(cst_processed['Eye'].value_counts())
print(f"\nArm distribution:")
print(cst_processed['Arm'].value_counts())
```

```python
# Check data types and missing values for CST columns
cst_cols = ['Screen'] + [tp for tp in time_points[1:]]

print("CST values summary (non-missing counts):")
for col in cst_cols:
    if col in cst_processed.columns:
        non_missing = cst_processed[col].notna().sum()
        print(f"  {col}: {non_missing}/{len(cst_processed)} ({100*non_missing/len(cst_processed):.1f}%)")
```

```python
# Check CST value ranges (should be ~200-600 microns typically)
print("\nCST value ranges per time point:")
for col in cst_cols[:5]:  # Check first 5 time points
    if col in cst_processed.columns:
        values = pd.to_numeric(cst_processed[col], errors='coerce')
        print(f"  {col}: min={values.min():.0f}, max={values.max():.0f}, mean={values.mean():.1f}")
```

## 5. Compare with BCVA Processed Format

```python
# Load BCVA processed for comparison
bcva_processed = pd.read_csv(r'c:\Users\ronny\Documents\GitHub\Longitudinal Prediction\1 BCVA_processed.csv')
print("BCVA processed columns:")
print(bcva_processed.columns.tolist())
print(f"\nCST processed columns:")
print(cst_processed.columns.tolist())
```

```python
# Verify Patient_ID alignment
bcva_patients = set(bcva_processed['Patient_ID'])
cst_patients = set(cst_processed['Patient_ID'])

print(f"BCVA patients: {len(bcva_patients)}")
print(f"CST patients: {len(cst_patients)}")
print(f"Overlap: {len(bcva_patients & cst_patients)}")
print(f"\nIn BCVA but not CST: {bcva_patients - cst_patients}")
print(f"In CST but not BCVA: {cst_patients - bcva_patients}")
```
